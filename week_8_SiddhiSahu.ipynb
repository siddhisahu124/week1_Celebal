# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# 🛠️ TOOL 3: Word Counter

def count_words(text: str) -> int:
    """Count words in a sentence."""
    try:
        return len(text.split())
    except Exception:
        return 0
````

````python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
````

````python
# Tool 4 : Text Summarizer
def summarize_text(text):
    try:
        sentences = text.split(".")
        return sentences[0]
    except:
        return "Unable to summarize"
````

````python
# Tool 5 : Sentiment Analyzer
def sentiment_analysis(text):

    positive_words = [
        "good",
        "great",
        "excellent",
        "happy",
        "awesome"
        "nice"
        "mesmerizing"
    ]

    negative_words = [
        "bad",
        "sad",
        "terrible",
        "awful",
        "poor"
        "eww"
    ]

    text = text.lower()

    pos = sum(word in text for word in positive_words)
    neg = sum(word in text for word in negative_words)

    if pos > neg:
        return "Positive"

    elif neg > pos:
        return "Negative"

    else:
        return "Neutral"
````

````python
# Tool 6 : Human Name Generator
import random

def generate_name():
    names = [
        "Siddhi",
        "Arman",
        "Maan",
        "Si",
        "Ankit",
        "Rohan",
        "Sneha",
        "Arjun"
    ]

    return random.choice(names)
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION

def agent(query: str):

    try:

        query_lower = query.lower()

        logging.info(f"Received Query: {query}")

        # ==========================
        # Calculator Tool Routing
        # ==========================
        if (
            "calculate" in query_lower or
            "solve" in query_lower or
            "math" in query_lower
        ):

            expression = (
                query_lower
                .replace("calculate", "")
                .replace("solve", "")
                .replace("math", "")
                .strip()
            )

            result = calculator(expression)

            logging.info("Calculator Tool Used")

            return {
                "type": "calculation",
                "result": result
            }

        # ==========================
        # Keyword Extraction Routing
        # ==========================
        elif (
            "keyword" in query_lower or
            "keywords" in query_lower or
            "extract" in query_lower
        ):

            text = (
                query
                .replace("Extract keywords from", "")
                .replace("extract keywords from", "")
                .strip()
            )

            result = extract_keywords(text)

            logging.info("Keyword Extraction Tool Used")

            return {
                "type": "keywords",
                "result": result
            }

        # ==========================
        # Word Count Tool Routing
        # ==========================
        elif "count words" in query_lower:

            text = query_lower.replace("count words", "").strip()

            result = count_words(text)

            logging.info("Word Count Tool Used")

            return {
                "type": "word_count",
                "result": result
            }

        # ==========================
        # Summarizer Tool Routing
        # ==========================
        elif "summarize" in query_lower:

            text = query.replace("Summarize", "").replace("summarize", "").strip()

            result = summarize_text(text)

            logging.info("Summarizer Tool Used")

            return {
                "type": "summary",
                "result": result
            }

        # ==========================
        # Sentiment Analysis Routing
        # ==========================
        elif "sentiment" in query_lower:

            text = query.replace("Sentiment", "").replace("sentiment", "").strip()

            result = sentiment_analysis(text)

            logging.info("Sentiment Analysis Tool Used")

            return {
                "type": "sentiment",
                "result": result
            }

        # ==========================
        # Name Generator Tool Routing
        # ==========================
        elif (
            "human name" in query_lower or
            "random name" in query_lower or
            "person name" in query_lower
        ):

            result = generate_name()

            logging.info("Name Generator Tool Used")

            return {
                "type": "name_generation",
                "result": result
            }

        # ==========================
        # General Response
        # ==========================
        else:

            logging.info("General Query Route Used")

            return {
                "type": "general",
                "result": "This query does not require any specialized tool."
            }

    except Exception as e:

        logging.error(f"Error: {str(e)}")

        return {
            "type": "error",
            "result": str(e)
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    # Calculator Tool
    "Calculate 20 % 5",

    # Keyword Extraction Tool
    "Extract keywords from Artificial Intelligence is transforming industries",

    # Word Counter Tool
    "Count words Artificial Intelligence is transforming industries",

    # Summarizer Tool
    "Summarize Artificial Intelligence is transforming industries. It is changing the future of work.",

    # Sentiment Analysis Tool
    "Sentiment This internship is awesome and excellent",

    # Name Generator Tool
    "Give me a random human name",

    # General Query
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
